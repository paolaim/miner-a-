## 2) Descripción de los datos

a) Qué tipo de datos son (tablas, imágenes, encuestas, etc.)
-Gracias a "The FAOSTAT" la base de datos nos muestra con actualizaciones anuales, las estadisticas sobre la media del cambio de temperatura en cada país. Esta posee los datos desde el año de 1961 hasta el 2019, de forma mensual, estacional y anual, utilizando de base el periodo de 1951 a 1980

b) Descripción de las columnas- (Que datos se describen en cada columna. Mediciones, tiempo, texto largo, etc.)
- Area Code: Entero, numero único asignado para cada región de la cual se tengan datos.
- Area: Texto, nombre en inglés de cada región.
- Months Code: Entero, número único asignado para cada mes, grupo o año.
- Months: Texto, nombre en inglés de los meses, grupo de meses y año meteorológico.
- Element Code: Entero, 7271 número asignado al cambio de temperatura o 6078 desviación estandar en su correspondiente tiempo.
- Element: Texto, cambio de temperatura o desviación estandar en el mes, grupo o año.
- Unit: Texto, tipo de grados en el que se midio la temperatura a nivel mundial.
- Y1961...Y2019: Flotante, valor total del cambio de temperatura o desviación estandar de manera anual comparado con la base de 1951 a 1980
